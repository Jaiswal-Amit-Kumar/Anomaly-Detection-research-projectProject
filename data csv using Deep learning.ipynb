```python
df = pd.read_csv('/content/drive/MyDrive/Datasets/pap files/data.csv')
```

```python
df.head(100)
```

```python
df = df.dropna()
```

```python
df = df.drop(['Unnamed: 0.1' , 'Unnamed: 0' , 'uid'] , axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['originh'] = le.fit_transform(df['originh'])
df['responh'] = le.fit_transform(df['responh'])
df['traffic_category'] = le.fit_transform(df['traffic_category'])
df.head()
```

```python
df.isnull().sum()
```

```python
x = df.drop(['Label'] , axis = 1)
y = df['Label']
```

```python
from sklearn.preprocessing import MinMaxScaler

mms = MinMaxScaler()
x_scaled = mms.fit_transform(x)
df1 = pd.DataFrame(x_scaled)

df1.head()
```

```python
from sklearn.decomposition import PCA

pcaa = PCA(n_components = 4)
x_pcaa = pcaa.fit_transform(x_scaled)
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test = train_test_split(x_pcaa , y , test_size = 0.2 , random_state = 42)
```

```python

# Model 1
start_time_model1 = time.perf_counter()


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Conv1D, GlobalMaxPooling1D, Dense

# Assuming X_train and y_train are your pre-encoded numerical features and labels

# Define the CNN model
model = Sequential()
model.add(Embedding(input_dim=1000, output_dim=64, input_length=x_train.shape[1]))
model.add(Conv1D(128, 3, activation='relu'))
model.add(GlobalMaxPooling1D())
model.add(Dense(3, activation='softmax'))  # Adjust units based on the number of classes

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test)
print("Accuracy:", accuracy)
print("Loss:", loss)

end_time_model1 = time.perf_counter()
elapsed_time_model1 = end_time_model1 - start_time_model1
print("Model 1 Execution Time:", elapsed_time_model1, "seconds")
```

```python

# Model 2
start_time_model2 = time.perf_counter()

from tensorflow.keras.layers import Dense

# Assuming X_train and y_train are your pre-encoded numerical features and labels

# Define the ANN model
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(x_train.shape[1],)))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))  # Adjust units based on the number of classes

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test)
print("Accuracy:", accuracy)
print("Loss:", loss)

end_time_model2 = time.perf_counter()
elapsed_time_model2 = end_time_model2 - start_time_model2
print("Model 2 Execution Time:", elapsed_time_model2, "seconds")
```

```python

# Model 3
start_time_model3 = time.perf_counter()

from tensorflow.keras.layers import SimpleRNN, Dense

# Assuming X_train and y_train are your pre-encoded numerical features and labels

# Define the RNN model
model = Sequential()
model.add(SimpleRNN(64, input_shape=(x_train.shape[1], 1), activation='relu'))  # Adjust units based on your data
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))  # Adjust units based on the number of classes

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Reshape X_train if needed (add a new axis for the time step)
#X_train = x_train.reshape(x_train.shape[0], x_train.shape[1], 1)

# Train the model
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# Reshape X_test if needed (add a new axis for the time step)
#X_test = x_test.reshape(x_test.shape[0], x_test.shape[1], 1)

# Evaluate the model
loss , accuracy = model.evaluate(x_test, y_test)
print("Accuracy:", accuracy)
print("Loss:", loss)

end_time_model3 = time.perf_counter()
elapsed_time_model3 = end_time_model3 - start_time_model3
print("Model 3 Execution Time:", elapsed_time_model3, "seconds")
```
